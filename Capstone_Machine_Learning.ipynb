# IMPORT LIBRARIES

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
from sklearn.metrics import jaccard_score
from sklearn.metrics import f1_score
from sklearn.metrics import log_loss
import itertools
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter
import pandas as pd
import numpy as np
import matplotlib.ticker as ticker
from sklearn import preprocessing
%matplotlib inline
```

# IMPORT DATA INTO PANDAS DATAFRAME

```python
#Training data set
df_train = pd.read_csv('loan_train.csv')
df_train.head()
```

```python
# Test data set
df_test = pd.read_csv('loan_test.csv')
df_test.head()
```

# VIEWING CATEGORIES FOR CATEGORICAL VARIABLES

```python
df_train['loan_status']=pd.Categorical(df_train['loan_status'])
print(df_train['loan_status'].cat.categories)
```

```python
df_train['education']=pd.Categorical(df_train['education'])
print(df_train['education'].cat.categories)
```

```python
df_train['Gender']=pd.Categorical(df_train['Gender'])
print(df_train['Gender'].cat.categories)
```

```python
df_test['loan_status']=pd.Categorical(df_test['loan_status'])
print(df_test['loan_status'].cat.categories)
```

```python
df_test['loan_status']=pd.Categorical(df_test['loan_status'])
print(df_test['loan_status'].cat.categories)
```

```python
df_test['loan_status']=pd.Categorical(df_test['loan_status'])
print(df_test['loan_status'].cat.categories)
```

# DATA CLEANING

# CONVERING CATEGORICAL VARIBLES TO NUMERICAL VARIABLES

```python
df_train['Gender'].replace(to_replace=['male','female'], value=[0,1],inplace=True)
```

```python
df_test['Gender'].replace(to_replace=['male','female'], value=[0,1],inplace=True)
```

```python
df_train['loan_status'].replace(to_replace=['PAIDOFF','COLLECTION'], value=[0,1],inplace=True)
```

```python
df_test['loan_status'].replace(to_replace=['PAIDOFF','COLLECTION'], value=[0,1],inplace=True)
```

# CONVERTING ORDINAL VARIABLES TO NUMERICAL

```python
Feature_train = df_train[['Principal','terms','age','Gender']]
Feature_train = pd.concat([Feature_train,pd.get_dummies(df_train['education'])], axis=1)
Feature_train.drop(['Master or Above'], axis = 1,inplace=True)
Feature_train.head()
```

```python
Feature_test = df_test[['Principal','terms','age','Gender']]
Feature_test = pd.concat([Feature_test,pd.get_dummies(df_test['education'])], axis=1)
Feature_test.drop(['Master or Above'], axis = 1,inplace=True)
Feature_test.head()
```

# FEATURE SELECTION

```python
x_train = Feature_train
```

```python
from sklearn import preprocessing
x_test = Feature_test
```

```python
y_train = df_train['loan_status'].values
```

```python
y_test = df_test['loan_status'].values
```

# APPLYING THE MACHINE LEARNING MODELS

# k-Nearest Neighbour

```python
from sklearn.neighbors import KNeighborsClassifier
k = 3
kNN_model = KNeighborsClassifier(n_neighbors=k).fit(x_train,y_train)
knn_yhat = kNN_model.predict(x_test)
print("KNN Jaccard index: %.2f" % jaccard_score(y_test, knn_yhat))
print("KNN F1-score: %.2f" % f1_score(y_test, knn_yhat, average='weighted') )
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
DT_model = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
DT_model.fit(x_train,y_train)
DT_yhat = DT_model.predict(x_test)
print("DT Jaccard index: %.2f" % jaccard_score(y_test, DT_yhat))
print("DT F1-score: %.2f" % f1_score(y_test, DT_yhat, average='weighted') )
```

# Support Vector Machine

```python
from sklearn import svm
SVM_model = svm.SVC()
SVM_model.fit(x_train, y_train)
SVM_yhat = SVM_model.predict(x_test)
print("SVM Jaccard index: %.2f" % jaccard_score(y_test, SVM_yhat))
print("SVM F1-score: %.2f" % f1_score(y_test, SVM_yhat, average='weighted') )
```

# LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
LG_model = LogisticRegression(random_state=0).fit(x_train, y_train)
LG_yhat = LG_model.predict(x_test)
LG_yhat_prob = LG_model.predict_proba(x_test)
print("LG Jaccard index: %.2f" % jaccard_score(y_test, LG_yhat))
print("LG F1-score: %.2f" % f1_score(y_test, LG_yhat, average='weighted') )
print("LR LogLoss: %.2f" % log_loss(y_test, LG_yhat_prob))
```
